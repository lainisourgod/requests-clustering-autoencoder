```python
def create_dataset(filename):
    samples = pd.DataFrame(columns=['id', 'class', 'vector'])
    i = 0
    for line in open(filename, 'r'):
        # Remove '\n' at the end of line
        line = line.rstrip()
        line = line.split(': ', maxsplit=1)

        if line[0] == 'Start - Id':
            lines = []
            sample = {'id': i}

        elif line[0] == 'End - Id':
            sample['class'] = lines[1][1]
            request_line = lines[2][0].split()
            if request_line[0] == 'GET':
                sample['vector'] = request_line[1]
            elif request_line[0] == 'POST' or request_line[0] == 'PUT':
                sample['vector'] = lines[-2][0]
            
            samples = samples.append([sample])
            if i % 1000 == 0:
                print(f'{i} requests processed')
            i += 1

        lines.append(line)

    return samples.reset_index(drop=True)
```

```python
%%time
create_dataset('./dataset_ecml_pkdd_train_test/xml_train.txt').to_csv('ecml_train.csv', index=False)
```
